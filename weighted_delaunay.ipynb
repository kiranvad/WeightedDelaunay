# Computes Weighted Delaunay Triangulation


```python
import numpy as np
from scipy.spatial import ConvexHull
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
def WeightedDelaunay(points,weights):

    num, dim = np.shape(points)

    lifted = np.zeros((num,dim+1))

    for i in range(num):
        p = points[i,:]
        lifted[i,:] = np.append(p,np.sum(p**2) - weights[i]**2)
    pinf = np.append(np.zeros((1,dim)),1e10);    
    lifted = np.vstack((lifted, pinf))
    hull = ConvexHull(lifted)
    delaunay = []
    for simplex in hull.simplices:
        if num not in simplex:
            delaunay.append(simplex.tolist())
            
    return delaunay
```

```python
mpl.rcParams['figure.dpi'] = 400
from matplotlib import rc
rc('text', usetex=True)

num_points = 10
points = np.random.rand(num_points, 2)
weights = np.random.rand(num_points,1)
delaunay = WeightedDelaunay(points,weights)
delaunay = np.asarray(delaunay)
print(delaunay)

fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.triplot(points[:,0], points[:,1], delaunay, 'b-', lw=1.0)
ax.scatter(points[:,0], points[:,1],s = 500*(weights/np.max(weights)),c = 'b')
counter = 1
for x,y in zip(points[:,0],points[:,1]):

    label = "{}".format(counter)

    ax.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,0), # distance from text to points (x,y)
                 ha='right') # horizontal alignment can be left, right or center
    counter = counter+1
ax.set_title('Weighted Delaunay Triangulation')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```
